Are there any missing values in the dataset? If so, how can they be handled?


```python
df.isnull().sum()
```

There is no any missing values in the dataset, if there were any missing values in the dataset, I would have drop the columns which contained rows with all null values by the function df.dropna(how = 'all'). I would also have replaced any specific null values by the mode, median or mean of the column for eg. df['column'].fillna(df.mean) also for categorical data I would have replaced with 'Not_given' or something like that so that, I can compare all the values in the dataset for eg: df['Gender'].fillna('Not given').
we also can drop any rows which contain one or more null values by df.dropna(how = 'any')

```python
df.dropna(how = 'all')
df['Salary'].fillna(df.mean)
df['Gender'].fillna('Not Given')
```
